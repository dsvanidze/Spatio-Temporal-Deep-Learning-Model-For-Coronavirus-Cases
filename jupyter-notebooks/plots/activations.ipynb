```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras import backend as K
from keras import activations
import math

points = 100
zeros = np.zeros((points,1))

df = pd.DataFrame({"activation": np.linspace(-10,10,points)})
df["sigmoid"] = K.eval(activations.sigmoid(df["activation"]))
df["tanh"] = K.eval(activations.tanh(df["activation"]))
df["relu"] = K.eval(activations.relu(df["activation"]))
```

```python
fig = plt.figure(figsize=(12,4))
ax = fig.add_subplot(1, 3, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position(('axes', 0.03))
# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_yticks(np.arange(0.1, 1.1, step=0.1))
ax.set_title("Sigmoid", fontsize=20, pad=20)
fig.tight_layout()
plt.plot(df["activation"], df["sigmoid"], "#2a3aa3", linewidth=3)

#############
ax = fig.add_subplot(1, 3, 2)
# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_yticks([*np.arange(0.2, 1.2, step=0.2), *np.arange(-1, 0, step=0.2)])
ax.set_title("TanH", fontsize=20, pad=20)
fig.tight_layout()
plt.plot(df["activation"], df["tanh"], "#FFA500", linewidth=3)

#############
ax = fig.add_subplot(1, 3, 3)
# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position(('axes', 0.03))
# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_yticks(np.arange(1, 11, step=1))
ax.set_title("ReLU", fontsize=20, pad=20)
fig.tight_layout()
plt.plot(df["activation"], df["relu"], "#60e65c", linewidth=3)
```
